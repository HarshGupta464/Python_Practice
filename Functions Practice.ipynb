```python
# Local Variable:
x= 24
print("First x value:", x)
def func():
    x=25
    return("Value of x in function:", x)
print(func())
print("Last value of x:", x)
```

```python
# Global Variable:
x=24
print("First x value:", x)
def func():
    global x
    x=25
    return("Value of x in function:", x)
print(func())
print("Last value of x:", x)
```

```python
# create a list and square there values between 1 and 30
def squares():
    l= []
    for i in range (1,31):
        l.append(i**2)
    return l
print(squares())
```

```python
# check prime number:
def prime(num):
    if num==1:
        print("Not a prime number")
    elif num==2:
        print("Prime Number")
    else:
        for i in range(2,num):
            if num%i==0:
                print("Not a Prime number")
                break
        else:
            print("Prime Number")
prime(47)
```

```python
# Function to sum all nos in a list
# solution 1
def add(nums):
    total=0
    for i in nums:
        total += i
    return total
print(add([6,9,3,4,2]))
```

```python
# solution 2 using recursion
def adds(nums):
    if len(nums)==1:
        return(nums[0])
    else:
        return (nums[0] + adds(nums[1:]))
print(adds([9,3,5,7,2,8]))
```

```python
# Function using recursion to the value of the fibonacci series at a particular index
def fib(num):
    if num == 1:
        return (0)
    if num == 2:
        return (1)
    else:
        return (fib(num-1)+fib(num-2))
print(fib(8))
```
