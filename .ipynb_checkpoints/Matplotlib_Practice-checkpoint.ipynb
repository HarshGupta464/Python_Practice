```python
from matplotlib import pyplot as plt
import numpy as np
x=[1,2,3,4,5,6,7,8,9]
y=[10,2,3,6,4,9,7,4,2]
#plotting graph with marker
plt.plot(x,y,"s")       #s,p,d,P
plt.show()
```

```python
#dotted plot with line
plt.plot(x,y,'s:b')
plt.show()
```

```python
# #Make solid line
plt.plot(x,y,'s--b')
plt.show()
```

```python
#Change Marker size:
plt.plot(x,y,'s-.b', ms=10)     #ms: Marker size
plt.show()
```

```python
#change the edge and fill color of marker
a=np.array([0,6,20,40])
b=np.array([3,8,1,10])
plt.plot(a,b, 's-.b', ms=12, mec="r", mfc="g")        #mec: marker edge color, #mfc: marker face color
plt.show()
```

```python
#Change the size of the plotline
plt.plot(a,b, ls="-", color="r", linewidth= "10")
plt.show()
```
