```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("hotstar.csv")
df
```

```python
df.head()
```

```python
df.tail(3)
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df['running_time']=df['running_time'].fillna(df['running_time'].median())
```

```python
df.duplicated().sum()
```

```python
sns.countplot(x=df['type'], data = df, color='coral') #to count the no of times each vale in type colm is appearing
plt.title("count of movies and tv shows")
plt.show()
```

```python
gb=df.groupby('type').agg({"hotstar_id":"count"})
gb
```

```python
plt.figure(figsize = (10,4))
sns.countplot(x=df['genre'],data = df, color='navy')
plt.title("Count of Genres")
plt.xticks(rotation = 90)
plt.show()
```

```python
gb=df.groupby("age_rating").agg({"hotstar_id":"count"})
gb
```

```python
plt.pie(gb['hotstar_id'], labels = gb.index, autopct = "%.2f%%")
plt.show()
```

```python
gb=df.groupby("year").agg({"hotstar_id":"count"})
gb
```

```python
plt.figure(figsize=(10, 4))
sns.lineplot(x=gb.index, y=gb['hotstar_id'], color='navy')
plt.title("Count of Hotstar IDs by Year")
plt.xlabel("Year")
plt.ylabel("Count of Hotstar IDs")
plt.show()
```
