```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df=pd.read_csv("Diwali Sales Data.csv", encoding="unicode_escape")
```

```python
df.info()
```

```python
df.shape
```

```python
df.head(10)
```

```python
df.drop(['Status', 'unnamed1'], axis=1, inplace = True)
```

```python
df.isnull().sum()
```

```python
df=df.dropna()
```

```python
df['Amount'].dtypes
```

```python
df['Amount']=df['Amount'].astype('int')
```

```python
df.describe()
```

```python
df[['Age', 'Orders', 'Amount']].describe()
```

```python
df.info()
```

```python
df.columns
```

```python
plt.figure(figsize=(8,6))
ax=sns.countplot(x='Gender', data=df, hue='Gender', palette=['pink', 'blue'])
for bars in ax.containers:
    ax.bar_label(bars)
plt.title("Total order count by Gender")
```

```python
df.groupby(['Gender'], as_index= False)['Amount'].sum().sort_values(by='Amount', ascending= False)
```

```python
plt.figure(figsize=(8,6))
sales_gen= df.groupby(['Gender'], as_index= False)['Amount'].sum().sort_values(by='Amount', ascending= False)
sns.barplot(x='Gender', y='Amount', data=sales_gen, hue='Gender', palette={'M': '#1f77b4', 'F': '#ff69b4'}, legend=False)
plt.title("Total Sale amount by Gender")
```

From the above two bar graphs, we can see that most buyers are females and even the purchasing power of females are greater than males

```python
plt.figure(figsize=(11,6))
ax=sns.countplot(x= 'Age Group', data=df, hue='Gender')
# plt.xticks(rotation=90)
plt.title("Total order count by Age Group")
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_age=df.groupby(['Age Group'], as_index= False)['Amount'].sum().sort_values(by='Amount', ascending= False)
# plt.figure(figsize=(10,6))
sns.barplot(x='Age Group', y= 'Amount', data= sales_age, color='cyan')
plt.title("Total sale amount by Age Group")
```

From the above 2 graphs, we can see that most of the buyers are of 26-35 age group females

```python
df.columns
```

```python
#Total orders from top  states
sales_state=df.groupby(['State'], as_index= False)['Orders'].sum().sort_values(by= 'Orders', ascending= False).head(10)
plt.xticks(rotation=90)
sns.barplot(x='State', y='Orders', data = sales_state, color= 'purple')
plt.title("Total orders from top 10 States")
plt.figure(figsize=(10,4))
```

```python
sales_state=df.groupby(['State'], as_index= False)['Amount'].sum().sort_values(by='Amount', ascending= False).head(10)
plt.xticks(rotation=90)
sns.barplot(x='State', y='Amount', data= sales_state, hue='State')
plt.title('Sales amount from top 10 states')
```

From above two charts we can see that most orders and the amount spent are from UP, Maharastra and Karnataka.
We can also see that Kerala has more orders than Haryana and Gujarat but spends less amount than those two states.
We can also see that Bihar has less orders than Gujarat but spends more amount than Gujarat.

```python
#sale analysis according to maritial status
df['Marital_Status'].value_counts().plot.pie(
    autopct='%1.1f%%',
    colors=sns.color_palette("Set2"),
    startangle=90,
    figsize=(5,5))
plt.title('Marital Status Distribution', fontsize=14, weight='bold')
plt.ylabel('')
plt.show()
```

```python
sales_marital= df.groupby(['Marital_Status'], as_index=False)['Amount'].sum().sort_values(by='Amount')
plt.figure(figsize=(5,5))
plt.pie(sales_marital['Amount'],
        labels=sales_marital['Marital_Status'],
        autopct='%1.1f%%',
        startangle=300,
        colors=sns.color_palette('coolwarm'))
plt.title('Sales Distribution by Marital Status', fontsize=14, weight='bold')
plt.show()
```

From the above two graphs we can see that unmarried people have spent more than married ones.

```python
plt.figure(figsize=(12,6))
order= df['Occupation'].value_counts().index
ax=sns.countplot(x='Occupation', data=df, hue='Occupation', order=order)
plt.xticks(rotation=90)
plt.title('Customer count by occupation', fontsize=14, weight='bold')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_occu= df.groupby(['Occupation'], as_index= False)['Amount'].sum().sort_values(by='Amount', ascending= False)
plt.figure(figsize=(12,6))
plt.xticks(rotation= 90)
sns.barplot(data=sales_occu, x= 'Occupation', y='Amount', hue= 'Occupation')
plt.title('Total sale amount by Occupation')
```

From the above two graphs we see that most of the buyers are from IT sector, Healthcare and Aviation sectors.

```python
order=df['Product_Category'].value_counts().index
plt.figure(figsize=(12,6))
plt.xticks(rotation=90)
sns.countplot(x='Product_Category', data=df, order=order, hue='Product_Category')
plt.title("Total orders by product category")
```

```python
# Analysis by product category
sales_product= df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending= False)
plt.figure(figsize=(12,6))
plt.xticks(rotation=90)
sns.barplot(data= sales_product, x='Product_Category', y= 'Amount', hue='Product_Category')
plt.title("Total Sale amount by Product Category")
```

From above two graphs we can see that most sold products are from food, Clothing & Apparel and Electronics and Gadgets
We can also see that Food has contributes more in total amount than Cothing & Apparel but are sold less than them

Conclusion:
It is concluded that Married woman of age group 26-35 from UP, Maharastra and Karnataka working in IT, Healthcare and Aviation sectors are more likely to buy products from Food, Clothing & Electronics category.
